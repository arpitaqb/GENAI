### ollama


```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embedding = (
    OllamaEmbeddings(model="llama3.2:1b") #by default it will use llama2
)
```

```python
embedding
```

```python
r1 = embedding.embed_documents(
    [
        "Alpha is the first letter of Greek alphabet",
        "Beta is the second letter of Greek alphabet",
    ]
)
```

```python
print(r1[0])
len(r1[0]) #llama3.2:1b provide 2048 dimention with respect to vectores
```

```python
embedding.embed_query("what is the second letter of greek alphabet")
```

```python
embedding = OllamaEmbeddings(model="mxbai-embed-large")
txt = "this is a test sentence."
query_result = embedding.embed_query(txt)
query_result
```
